```python
def make_null(matrix):
    rows_to_nullify, cols_to_nullify = np.random.randint(19, size=5), np.random.randint(15, size=5)
    matrix[rows_to_nullify, cols_to_nullify] = np.nan
    return matrix
A=A.astype(float)
make_null(A)
```

```python
def restore_null_values(matrix):
    column_means = np.nanmean(matrix, axis=1)
    null_indices = np.argwhere(np.isnan(matrix))
    for row_idx, col_idx in null_indices:
        matrix[row_idx, col_idx] = column_means[row_idx]
    return matrix

restore_null_values(A)
```

```python
def normalize(A):
    return  (A - A.min()) / (A.max() - A.min())

normalize(A)
```

```python
import matplotlib.pyplot as plt
from skimage import io

image_path = 'C:\\Users\\Sony\\Desktop\\Downloads\\owl.jpg'
image = io.imread(image_path)

plt.figure(figsize=(12, 6))
plt.imshow(image)
plt.title('Оригінальне зображення')
plt.axis('off')
```

```python
from skimage import exposure

min_value = 50
max_value = 200

image_contrast = exposure.rescale_intensity(image, in_range=(min_value, max_value))

plt.figure(figsize=(12, 6))
plt.imshow(image_contrast)
plt.title('Зображення зі зміненою контрастністю')
plt.axis('off')

plt.show()
```

```python
blue_channel = image[:, :, 0]
green_channel = image[:, :, 1]
red_channel = image[:, :, 2]

plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.imshow(blue_channel, cmap='Blues')
plt.title('Синій канал')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(green_channel, cmap='Greens')
plt.title('Зелений канал')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(red_channel, cmap='Reds')
plt.title('Червоний канал')
plt.axis('off')

plt.show()
```

```python
plt.figure(figsize=(6, 6))

plt.imshow(image, cmap='gray')
plt.title('Чорно-біле зображення')
plt.axis('off')

plt.show()
```

```python
from scipy.ndimage import gaussian_filter

blurred_image = gaussian_filter(image, sigma=(30,1,10))

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)

plt.imshow(image)
plt.title('Оригінальне зображення')
plt.axis('off')

plt.subplot(1, 2, 2)

plt.imshow(blurred_image)
plt.title('Розмите зображення')
plt.axis('off')

plt.show()
```

```python
from skimage import color
from skimage.filters import roberts, sobel

gray_image = color.rgb2gray(image)

edge_roberts = roberts(gray_image)

edge_sobel = sobel(gray_image)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(edge_roberts, cmap='gray')
plt.title('Roberts Edge Detection')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(edge_sobel, cmap='gray')
plt.title('Sobel Edge Detection')
plt.axis('off')

plt.show()
```

```python
from skimage import color, measure

gray_image = color.rgb2gray(image)

contours = measure.find_contours(gray_image, 0.6)

plt.figure(figsize=(12, 6))
plt.imshow(gray_image, cmap='gray')

for contour in contours:
    plt.plot(contour[:, 1], contour[:, 0], linewidth=1)

plt.title('Зображення з контурами')
plt.axis('off')
plt.show()
```
